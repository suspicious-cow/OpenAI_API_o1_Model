# OpenAI o1 Initial Logic Test

## LSAT question

Taken from https://phoenixlsattutor.wordpress.com/2011/09/12/logic-games-more-sequencing-practice-part-1/

```python
# Importing required libraries

# 'openai' is a module that provides access to OpenAI's API, 
# allowing interaction with language models like GPT.
from openai import OpenAI

# 'display' and 'Markdown' are used to display formatted markdown content
# in environments like Jupyter Notebooks.
from IPython.display import display, Markdown
```

```python
# Create an instance of the OpenAI client
client = OpenAI()

# The client sends a chat completion request using the OpenAI API.
# This uses the specified model ("o1-preview") and a list of messages.
completion = client.chat.completions.create(
    model="o1-preview",  # Specify the model to use
    messages=[
        # The user-provided message content contains a logic puzzle with questions.
        {"role": "user", "content": """
        A farmer harvests eight separate fields—G, H, J, K, L, M, P, 
        and T. Each field is harvested exactly once, and no two fields 
        are harvested simultaneously. Once the harvesting of a field 
        begins, no other fields are harvested until the harvesting of 
        that field is complete. The farmer harvests the fields in an 
        order consistent with the following conditions:

        Both P and G are harvested at some time before K.
        Both H and L are harvested at some time before J.
        K is harvested at some time before M but after L.
        T is harvested at some time before M.

        Questions:
        1. Which entities cannot be first?
        2. Which entities cannot be last?
        3. What’s the earliest possible place K can go?
        4. Is there a relationship between G & T?
        5. Is there a relationship between P & M?
        """}
    ],
    temperature=1,         # Controls randomness in the model's response (1 means high variability)
    top_p=1,                # Determines diversity via nucleus sampling (1 means the most diverse)
    frequency_penalty=0,    # No penalty for repeating frequent tokens
    presence_penalty=0,     # No penalty for introducing new topics
)

# 'completion' now holds the response from the OpenAI API.
```

```python
# Extract the content from the first choice of the completion response.
# 'completion.choices[0].message.content' refers to the model's response in the first message.
content = completion.choices[0].message.content

# Display the content as Markdown.
display(Markdown(content))
```
